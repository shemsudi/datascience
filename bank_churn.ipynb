import nessasary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

load the data

```python
data=pd.read_csv("train.csv")
data.head()
```

describing the data

```python
data.describe()
```

independent and dependent variable 

```python
data["Gender"]=data["Gender"].map({"Male": 1,"Female":0})
```

```python
X=data[["CreditScore","Age","Gender","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]
y=data["Exited"]
```

```python
X=np.array(X)
y=np.array(y)
```

```python
data.describe()
```

```python
X=sm.add_constant(X)
results=sm.OLS(y,X).fit()
results.summary()
```

exploring the data

```python
testData=pd.read_csv("test.csv")
```

```python
testData["Gender"]=testData["Gender"].map({"Male": 1,"Female":0})
```

```python
data1=testData[["CreditScore","Age","Gender","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]
data1=sm.add_constant(data1)
data1.head()
```

```python
predictions=results.predict(data1)
id=testData["id"]
```

```python
result_df = pd.DataFrame({"id": id, "Excited": predictions})
result_df.to_csv('submission.csv', index=False)
```
